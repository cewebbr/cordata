```python
import pandas as pd
import matplotlib.pyplot as pl
```

```python
df = pd.read_csv('../dados/backups/counts.csv', names=['uc', 'type', 'date', 'format', 'total_counts'])
df['date'] = pd.to_datetime(df['date'])
df['added'] = df['total_counts'].diff()
```

```python
pl.figure(figsize=(15,4))
df.set_index('date')['total_counts'].plot()
pl.axvline(pd.to_datetime('2025-12-20'))
pl.axvline(pd.to_datetime('2026-01-07'))
```

```python
df.loc[df['date'] > pd.to_datetime('2026-01-07'), 'added'].hist()
```
